python difference between raw_input and input  |  no colab não funciona o raw_input

Basically, the difference between raw_input and input is that the return type of raw_input is always string, while the return type of input need not be string only. Python will judge as to what data type will it fit the best. In case you have entered a number, it will take it as an integer.

```
## continue =   JUMPS TO THE "TOP" OF THE LOOP ANS STARTSS THE NEXT INTERACTION

while True:
   line = input ('>')
   if line == '#':
     continue
   if line == 'done':
     break
   print (line)
print ('Done')
```

```
largest_so_far = -1
print ('before', largest_so_far)
for the_num in [9, 41, 12, 3, 74, 15]:
  if the_num > largest_so_far:
    largest_so_far = the_num
  print (largest_so_far, the_num)

print ('after', largest_so_far)
```

```
smallest_so_far = None
print ('before')
for the_num in [9, 41, 12, 3, 74, 15]:
  if smallest_so_far is None:
    smallest_so_far = the_num
  elif the_num < smallest_so_far:
    smallest_so_far = the_num
  print (smallest_so_far, the_num)

print ('after', smallest_so_far)
```

Escreva um programa que solicite repetidamente a um usuário números inteiros até que o usuário digite 'concluído'.

Depois de inserir 'concluído', imprima o maior e o menor dos números. Se o usuário inserir qualquer coisa diferente de um número válido, pegue-o com um try / except e envie uma mensagem apropriada e ignore o número. Digite 7, 2, bob, 10 e 4 e corresponda à saída abaixo.

Invalid input
Maximum is 10
Minimum is 2

```
largest = None
smallest = None

while True :
  sval = input ("Enter a number: ")
  if sval == 'done':
    break
  try:
    fval = int (sval)
  except :
    print ('Invalid input')

  
  if largest is None:
    largest = fval
  elif largest < fval:
    largest =  fval

  if smallest is None:
    smallest = fval
  elif smallest > fval:
    smallest =  fval


print("Maximum is", largest)
print("Minimum is", smallest)
```
